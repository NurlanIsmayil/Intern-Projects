```python
data = data.drop(["Week of the month (first week, second, third, fourth or fifth week", "Day of the week (Monday to Friday)", "Urgent order","Fiscal sector orders","Orders from the traffic controller sector"], axis=1)
```

```python
X = data.drop("Non-urgent order", axis=1)
X.head()
```

```python
y = data["Non-urgent order"];y.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
y = y.values.reshape(-1,1)
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.30,random_state=100)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
lr.fit(X_train,y_train)
```

```python
pd.DataFrame(lr.intercept_, columns=['intersept'])
```

```python
pd.DataFrame(lr.coef_)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2 = r2_score(y_test,y_pred);r2
```

```python
mse = mean_squared_error(y_test,y_pred);mse
```

```python
import math
rmse = math.sqrt(mse);rmse
```
