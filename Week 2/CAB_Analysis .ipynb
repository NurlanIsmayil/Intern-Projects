```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

mask = (Data['Date_of_Travel'] >= 2016) & (Data['Date_of_Travel'] <= 2018)
Data = Data.loc[mask]
grouped = Data.groupby('Company')['Users'].sum().reset_index()

sns.barplot(x='Company', y='Users', data=grouped)
plt.xlabel('Company')
plt.ylabel('Total users')
plt.title('Cab users for companies Pinc Cab and Yellow Cab (2016-2018)')
plt.show()
```

```python
mask = (Data['Date_of_Travel'] >= 2016) & (Data['Date_of_Travel'] <= 2018)
Data = Data.loc[mask]
grouped = Data.groupby(['Date_of_Travel', 'Company'])['Users'].sum().reset_index()

plt.figure(figsize=(13, 6))
sns.lineplot(x='Date_of_Travel', y='Users', hue='Company', data=grouped)
plt.xlabel('Year')
plt.ylabel('Total users')
plt.title('Cab users for companies Pinc Cab and Yellow Cab (2016-2018)')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```
